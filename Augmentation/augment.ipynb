```python
train_data = get_json(f'./data/curated/train.json')
questions = get_json(f'./data/metadata/acceptable.json')

avg_count_correct = 0
avg_count_partial_correct = 0
avg_count_incorrect = 0

for question in questions:
    avg_count_correct += questions[question]['count']['correct']
    avg_count_partial_correct += questions[question]['count']['partially correct']
    avg_count_incorrect += questions[question]['count']['incorrect']

avg_count_correct = avg_count_correct/len(questions)
avg_count_partial_correct = avg_count_partial_correct/len(questions)
avg_count_incorrect = avg_count_incorrect/len(questions)
print('correct: ', round(avg_count_correct, 2))
print('partial: ',round(avg_count_partial_correct, 2))
print('incorrect: ',round(avg_count_incorrect, 2))
```
